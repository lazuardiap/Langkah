```python
# Load tf decision forests
import tensorflow_decision_forests as tfdf
import tensorflow as tf

# load training data
import pandas
train_df = pandas.read_csv("Training.csv")
test_df = pandas.read_csv("Testing.csv")
```

```python
train_df = train_df.rename(columns={'dischromic _patches' : 'dischromic_patches', 
                                    'foul_smell_of urine' : 'foul_smell_of_urine',
                                    'spotting_ urination' : 'spotting_urination',
                                    'toxic_look_(typhos)' : 'toxic_look_typhos'}, inplace=False)
test_df = test_df.rename(columns={'dischromic _patches':'dischromic_patches', 
                                  'foul_smell_of urine' : 'foul_smell_of_urine',
                                  'spotting_ urination' : 'spotting_urination',
                                  'toxic_look_(typhos)' : 'toxic_look_typhos'}, inplace=False)
```

```python
# Convert the pandas df into tf dataset
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label="prognosis")
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df, label="prognosis")
```

```python
# Train the model
model = tfdf.keras.RandomForestModel()
model.fit(train_ds)
```

```python
# Evaluating the model
model.compile(metrics=["accuracy"])
print(model.evaluate(test_ds))
```

```python
# Save the model
save_dir = "/content"
tf.saved_model.save(model, save_dir)
```

```python
# Convert it to tflite
converter = tf.lite.TFLiteConverter.from_saved_model("/content")
tflite_model = converter.convert()
tflite_model_file = 'converted_chatbot_model.tflite'

with open(tflite_model_file, 'wb') as f:
  f.write(tflite_model)
```
